```python
def d_ou(v1,v2):  #np.dot(v1,v2) == np.sum(v1*v2)
    v1=np.array(v1)
    v2=np.array(v2)
    return np.sqrt(np.dot((v1-v2),(v1-v2)))
    #return np.sqrt(np.sum((v1-v2)*(v1-v2)))
```

```python
from sklearn.datasets import fetch_lfw_people
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np
```

```python
faces=fetch_lfw_people(min_faces_per_person=60)
faces.data.shape
faces.images.shape
x=faces.data
```

```python
pca=PCA(150).fit(x)
x_dr=pca.transform(x)
x_dr.shape
```

```python
fig,axes=plt.subplots(4,8,figsize=[10,5],subplot_kw={"xticks":[],"yticks":[]})
for i,ax in enumerate(axes.flat):
    ax.imshow(faces.data[i,:].reshape(62,47),cmap='gray')
```

```python
x_inverse=pca.inverse_transform(x_dr)
fig,axes=plt.subplots(4,8,figsize=[10,5],subplot_kw={"xticks":[],"yticks":[]})
for i,ax in enumerate(axes.flat):
    ax.imshow(x_inverse[i,:].reshape(62,47),cmap='gray')
```

```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn import datasets
 
y = np.array([[1,5,6],[4,3,9],[4,2,9],[4,7,2]])  # 四行三列
def pca(X,k):
    n,m = X.shape
    mean = np.mean(X,axis=0)
    # print(mean)
    X_norm = X - mean
    # print(X_mean)
    # return None
    scatter_matrix = np.dot(X_norm.T,X_norm)
    eig_val, eig_vec = np.linalg.eig(scatter_matrix)
    #print(eig_val,"\n")
    #print(eig_vec,'\n')
    Ord_eig_val = np.flipud(np.argsort(eig_val))
    feature = []
    for i in range(k):
        feature.append(eig_vec[:,Ord_eig_val[i]])
    feature = np.array(feature)
    #print(feature)
 
    data = np.dot(X_norm,feature.T)
    return data
 
########################
 
#
 
print(pca(y,2),'\n')
 
PCA_1 = PCA(n_components=2)
print(PCA_1.fit_transform(y))
```
